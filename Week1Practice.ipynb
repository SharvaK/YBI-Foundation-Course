```python
x = [1,2,3,4,5,6,7]
y = [1,2,3,4,5,6,7]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.barh(x,y)
plt.show()
```

```python
z = [1,1,1,1,2,2,3,3,3,4,4,4,5,5,6,6,6,7,7]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(z)
plt.show()

```

```python
fig, ax = plt.subplots()
ax.pie(y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y , color = 'red', linestyle ='--', marker = 'v', linewidth = 4)
ax.set_title('Sales for FY 2021')
ax.set_xlabel('Months')
ax.set_ylabel('Sales')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```
