```python
error_rate = []

for i in range(1,40) :
  knncla = KNeighborsClassifier(n_neighbors=i)
  knncla.fit(X_train, y_train)
  pred_i = knncla.predict(X_test)
  error_rate.append(classification_report(y_test,pred_i))
```
